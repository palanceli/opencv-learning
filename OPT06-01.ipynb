# 基本绘制

```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## 线条

```python
img = np.zeros((300, 400, 3), np.uint8)

# 画线，参数为：起点、终点坐标、颜色、线宽
cv2.line(img, (0, 0), (200, 300), (255, 0, 0), 1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 矩形

```python
img = np.zeros((300, 400, 3), np.uint8)

# 矩形，参数为：左上角、右下角坐标、颜色、线宽
cv2.rectangle(img, (50, 100), (150, 250), (0, 255, 0), 2)
cv2.rectangle(img, (200, 50), (300, 250), (0, 255, 0), -1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 圆

```python
img = np.zeros((300, 400, 3), np.uint8)

# 圆，参数为：圆心坐标、半径、颜色、线宽
cv2.circle(img, (100, 100), 50, (0, 0, 255), 1)
cv2.circle(img, (250, 150), 80, (0, 0, 255), -1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 椭圆

```python
img = np.zeros((300, 400, 3), np.uint8)

# 椭圆，参数为：中心坐标、长短轴长度、起始旋转角度、起始绘制角度、中止绘制角度
cv2.ellipse(img, (100, 50), (80, 50), 0, 0, 180, (255, 0, 0), 2)
cv2.ellipse(img, (300, 50), (80, 50), 0, 60, 180, (255, 0, 0), -1)
cv2.ellipse(img, (300, 150), (80, 50), 30, 60, 180, (255, 0, 0), -1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 多边形

```python
img = np.zeros((300, 400, 3), np.uint8)

# 多边形，参数为：顶点坐标、是否封闭、颜色、线宽
pts = np.array([[50, 50], [150, 100], [200, 80], [150, 50]], np.int32)
cv2.polylines(img, [pts.reshape(-1, 1, 2)], False, (255, 255, 255), 1)

# 封闭多边形
pts = np.array([[250, 50], [300, 100], [350, 80], [320, 50]], np.int32)
cv2.polylines(img, [pts.reshape(-1, 1, 2)], True, (255, 255, 255), 1)

# 顶点坐标不加中括号，则只绘制顶点
pts = np.array([[120, 180], [150, 280], [200, 220], [250, 210]], np.int32)
cv2.polylines(img, pts.reshape(-1, 1, 2), True, (255, 255, 255), 2)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 文字

```python
img = np.zeros((300, 400, 3), np.uint8)

# 文字
font = cv2.FONT_HERSHEY_SIMPLEX                             # 文字
cv2.putText(img, 'OpenCV', (50, 50), font, 2, (255, 255, 255), 2, cv2.LINE_AA)
cv2.putText(img, 'OpenCV', (50, 200), font, 1, (255, 255, 255), 1, cv2.LINE_AA)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
