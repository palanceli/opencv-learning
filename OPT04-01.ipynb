# 图片操作

由于在jupyter中使用cv2的`imshow(...)`函数显示会导致kernel crash，目前还不知道怎么解决，所以显示部分均使用`matplotlib`。

```python
import os
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
%pylab inline
```

## 读入

```python
img = cv2.imread('images/SGDevX.jpg')
```

## 显示
但是彩色图片使用OpenCV加载时是BGR模式，但Matplotlib是RGB模式，所以要做一个转换：

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

## 保存

```python
if not os.path.exists('temp'):
    os.mkdir('temp')
cv2.imwrite('temp/SGDevX.jpg', img)
```
